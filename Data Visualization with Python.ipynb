```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.dates as dates

# Loading the dataset

df = pd.read_csv(r"C:\Users\User\Downloads\retail_sales_dataset.csv")

df['Date'] = pd.to_datetime(df['Date'], format= '%Y-%m-%d')

df
```

```python
# visualizing the category of products based on the quantity

plt.figure(figsize=(10,6))
df.groupby('Product Category')['Quantity'].sum().plot(kind='bar', color='blue')
plt.title('Quantity by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Quantity')
plt.rcParams['axes.facecolor']='coral'
plt.show()
```

```python
# Using line graph to show the total quantity of goods sold over the months

plt.figure(figsize=(10,6))
df.groupby(df['Date'].dt.to_period('M'))['Quantity'].sum().plot(kind='line', color='indigo')
plt.title('Quantity by Date')
plt.xlabel('Date')
plt.ylabel('Quantity')
plt.rcParams['axes.facecolor']='lightcyan'
plt.show()
```

```python
# Showing the different age distribution across different product categories

g = sns.FacetGrid(df, col = 'Product Category', height=4, aspect=1.5)
g.map(sns.histplot, 'Age', bins=20)
plt.show()
```

```python
# Showing the count of gender in relation to each product category

plt.figure(figsize=(10,6))
sns.countplot(x = 'Product Category', hue = 'Gender', data = df)
plt.title('Count by Gender and Product Category')
plt.xlabel('Product Category')
plt.ylabel('Count')
plt.legend(title = 'Gender')
plt.show()
```

```python
# Showing the total amount of goods sold under each product category

plt.figure(figsize=(10,6))
df.groupby('Product Category')['Total Amount'].sum().plot(kind='bar', color='Blue')
plt.title('Total Amount by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Amount')
plt.rcParams['axes.facecolor']='coral'
plt.show()
```

```python
# Showing the trends of total amount of goods sold in each month

plt.figure(figsize=(10,6))
df.groupby(df['Date'].dt.to_period('M'))['Total Amount'].sum().plot(kind='line', color='indigo')
plt.title('Total Amount by Date')
plt.xlabel('Date')
plt.ylabel('Total Amount')
plt.rcParams['axes.facecolor']='coral'
plt.show()
```

```python
# Showing the total amount of goods sold by gender

plt.figure(figsize=(10,6))
df.groupby('Gender')['Total Amount'].sum().plot(kind='pie', autopct = '%1.1f%%', colormap= 'OrRd')
plt.title('Total Amount by Gender')
plt.xlabel('Gender')
plt.ylabel('Total Amount')
plt.rcParams['axes.facecolor']='blue'
plt.show()
```
